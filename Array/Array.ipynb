### 3. Create a list of all odd numbers between 1 and a max number. Max number is something you need to take from a user using input() function

```python
def odd_number(max_number):
    number =[]
    for no in range(1, max_number + 1):
        if no % 2 != 0:
            number.append(no)

    return number
    
output = odd_number(5)
print(output)
```

```python
# 2. Remove duplicates from a list
lst = [1,2,4,5,3,2,1]

# # using set 
# output = list(set(lst))
# print(output)

# using for loop
def remove_duplicate(input):
    output = []
    for i in input:
        if i not in output:
            output.append(i)
    return output

print(remove_duplicate(lst)) 


```

```python
# 4.Flatten a list
nested_list = [[1, 2], [3, 4], [5, 6]]

def flatten(input):
    output = []
    for i in input:
        if type(i) is list:
            output.extend(flatten(i))
        else:
            output.append(i)
    return output

print(flatten(nested_list))         
```

```python
# 6.Group anagrams
words = ["cat", "dog", "tac", "god", "act"]
# output =[["cat","tac","act"], ["dog","god"]]

def group_anagrams(input):
    d = {}
    for word in input:
        key = "".join(sorted(word))
        if key not in d:
            d[key] = []
        d[key].append(word)  
    output = [v for v in d.values()]      
    return output

print(group_anagrams(words))  
                

```

```python
# 7. We want to get employees and their salary where salary is greater than average salary of the company
emp_list = [('Ankit',10000), ('Rahul', 12000), ('Sumit',14000), ('Dheeraj', 21000), ('Pavan',11000), ('Mohit', 13000)]

def employees(input):
    sum = 0
    output = []
    count = len(input)
    for name, salary in input:
        sum = salary + sum
    avg = sum/count  

    for name , salary in input:
        if salary > avg:
            output.append(name)
    return output   

print(employees(emp_list))
```
