```python
# 5. Check a string is palindrome or not
# A palindrome is a string that reads the same forwards and backwards.
# example madam, racecar, level...

def palindrome(input):
    if input[0::] == input[::-1]:
        return(f"{input} is a palindrome")
    else:
        return(f"{input} is not a palindrome")    
    
print(palindrome('madam'))   
print(palindrome("waqar")) 
```

```python
# 6. Check if two strings are anagrams
# An anagram is when two strings contain the exact same characters but in a different order.,

word ="listen"
word1 = "silent"

def anagrams(input, input1):      # string
    if sorted(input) == sorted(input1):
        return(f"{input} and {input1} are anagrams")
    return(f"{input} and {input1} are not anagrams")

print(anagrams(word, word1))

```

```python
# 9. Given a string s and an integer k, return the maximum number of vowel letters in any substring of s with length k.
# LEETCODE 1456
s = "abciiidef"
k = 3

# using sliding window 
vowel = set('aeiou')
count = sum(1 for i in s[:k] if i in vowel)
max_count = count 

for i in range(k, len(s)):
    if s[i-k] in vowel:
        count -=1
    if s[i] in vowel:
        count +=1
        max_count = max(count,max_count)
print(max_count)        
```
