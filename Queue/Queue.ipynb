```python
# 1. Implement a queue using deque
from collections import deque
q = deque() # empty queue

# insert value in queue
q.append(1)
q.append(2)
q.append(3)
q.append(4)

# checking if the queue is empty or not 
if not q:
    print('Queue is empty')
else:
    print('Queue is not empty')   

# Checking the length of the queue
print(len(q))     

# removing the element from queue
q.popleft()


print(q)
```
