```python
# 1. Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.
# LEETCODE 20
# s = "()[]{}"
s = "(]"

def isvalid(s):
    stack = []
    if len(s) % 2 == 1 or len(s) == 0:
        return False
    for i in range(len(s)):
        if s[i] == '(' or s[i] == '{' or s[i] == '[':
            stack.append(s[i])
        else:
            if s[i] == ')' and stack.pop() == '(':
                return True
            if s[i] == ']' and stack.pop() == '[':
                return True 
            if s[i] == '}' and stack.pop() == '{':
                return True
            else:
                return False 
            
print(isvalid(s))            
    
            

    
```

```python
# 2. Minimum Add to Make Parentheses Valid
# LEETCODE 921

# s = "())"
s = "((("

def validparenthesis(s):
    stack = []
    count = 0
    if len(s) == 0:
        return 0
    for i in range(len(s)):
        if s[i] == '(':
            stack.append(s[i])
        else:
            if stack:
                stack.pop()
            else:
                count +=1
    return count + len(stack)                
                 

print(validparenthesis(s))   
```
