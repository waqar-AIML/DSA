## Level-2

```python
# Square pattern 
n = 5
for i in range(n):
    for j in range(n):
        print("*", end=" ")
    print()        
```

```python
# Number Square pattern 
n = 5
for i in range(n):
    for j in range(n):
        print(j+1, end=" ")
    print()    
```

## Level-3

```python
# Right Triangle star
n = 5
for i in range(1, n+1):
    print("*" * i)
```

```python
# Number Right Triangle star
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()    
```

```python
# Repeated Number Triangle
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(i, end = ' ')
    print()    
```

## Level-4

```python
# Reverse pattern 
n = 5

for i in range(n, 0, -1):
    print("*" * i)   
```

```python
# Reverse Number triangle
n = 5
for i in range(n,0,-1):
    for j in range(1, i+1):
        print(j, end = " ")
    print()    
```

## Level 5

```python
# Star Pyramid
n = 5
for i in range(1, n+1):
    for space in range (1, n - i + 1):
        print(end = " ")
    for star in range(1, i+1):
        print("*" , end = " ")
    print()        
```
